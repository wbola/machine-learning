```python
# unsupervised learning

import pandas as pd
import matplotlib.pyplot as plt

points  = pd.read_csv('points.csv', sep=' ', header=None)

xs = points[0]
ys = points[1]

points.plot(kind='scatter', x=0, y=1)

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)

kmeans.fit(points)

clusters = kmeans.predict(points)

plt.scatter(x=points[0], y=points[1], c=clusters)
plt.show()

for idx, centroid in enumerate(kmeans.cluster_centers_):
    print("Claster ID: {}\tX: {}\tY:{}".format(idx, centroid[0], centroid[1]))
    
# the number of clusters is equal to 2
kmeans1 = KMeans(n_clusters=2)
kmeans1.fit(points)
clusters1 = kmeans1.predict(points)
plt.scatter(x=points[0], y=points[1], c=clusters1)
plt.show()
for idx, centroid in enumerate(kmeans1.cluster_centers_):
    print("Claster ID: {}\tX: {}\tY:{}".format(idx, centroid[0], centroid[1]))

# the number of clusters is equal to 4
kmeans1 = KMeans(n_clusters=4)
kmeans1.fit(points)
clusters1 = kmeans1.predict(points)
plt.scatter(x=points[0], y=points[1], c=clusters1)
plt.show()
for idx, centroid in enumerate(kmeans1.cluster_centers_):
    print("Claster ID: {}\tX: {}\tY:{}".format(idx, centroid[0], centroid[1]))
    
n_clusters = [1, 2, 3, 4, 5]
inertias = []

for n_cluster in n_clusters:
    model = KMeans(n_clusters=n_cluster)
    model.fit(points)
    inertias.append(model.inertia_)

plt.plot(n_clusters, inertias)
plt.show()

from sklearn.cluster import DBSCAN

db = DBSCAN(eps=130, min_samples=10)
labels = db.fit_predict(points)
labels

plt.scatter(x=points[0], y=points[1], c=labels)
plt.show()

points=(points-points.mean())/points.var()

kmeans3 = KMeans(n_clusters=3)
kmeans3.fit(points)
clusters3 = kmeans3.predict(points)
plt.scatter(x=points[0], y=points[1], c=clusters3)
plt.show()
for idx, centroid in enumerate(kmeans3.cluster_centers_):
    print("Claster ID: {}\tX: {}\tY:{}".format(idx, centroid[0], centroid[1]))

db1 = DBSCAN(eps=130, min_samples=10)
labels1 = db1.fit_predict(points)
plt.scatter(x=points[0], y=points[1], c=labels1)
plt.show()

# dimension reduction

from sklearn.decomposition import PCA

from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original')
X = mnist.data.astype('float64')
y = mnist.target

pca = PCA(n_components=10)
pca.fit(X)

mnist_pca = pca.transform(X)

mnist = KMeans(n_clusters=10)
mnist.fit(mnist_pca)
mnist_clasters = mnist.predict(mnist_pca)

pd.crosstab(mnist_clasters,y)

a0=pd.crosstab(mnist_clasters,y).values[0]
a1=pd.crosstab(mnist_clasters,y).values[1]
a2=pd.crosstab(mnist_clasters,y).values[2]
a3=pd.crosstab(mnist_clasters,y).values[3]
a4=pd.crosstab(mnist_clasters,y).values[4]
a5=pd.crosstab(mnist_clasters,y).values[5]
a6=pd.crosstab(mnist_clasters,y).values[6]
a7=pd.crosstab(mnist_clasters,y).values[7]
a8=pd.crosstab(mnist_clasters,y).values[8]
a9=pd.crosstab(mnist_clasters,y).values[9]
s=a2[0]+a9[1]+a3[2]+a7[3]+a4[4]+a7[5]+a5[6]+a1[7]+a6[8]+a4[9]
print(s)

b=s/len(mnist_clasters)
print(b)
```
